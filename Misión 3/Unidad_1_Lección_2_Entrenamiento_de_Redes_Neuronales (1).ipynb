# Programación de redes neuronales

Paso 2: Importando el conjunto de datos

```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
# Crear los datos artificiales
X, Y = make_circles(n_samples=5000, noise=0.05, factor=0.5)

# Resolución del mapa de predicción.
res = 100

# Coordenadas del mapa de predicción
_x0 = np.linspace(-1.5, 1.5, res)
_x1 = np.linspace(-1.5, 1.5, res)

# input con cada combo de coordenadas
_pX = np.array(np.meshgrid(_x0, _x1)).T.reshape(-1, 2)

# Objeto vacio a 0.5
_pY = np.zeros((res, res)) + 0.5

```

```python
# Visualización del mapa de predicción.
plt.figure(figsize=(8, 8))
plt.pcolormesh(_x0, _x1, _pY, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='coolwarm', vmin=0, vmax=1)
plt.show
```

```python
# Visualización de la nube de datos
plt.scatter(X[Y == 0,0], X[Y == 0,1], c="skyblue")
plt.scatter(X[Y == 1,0], X[Y == 1,1], c="salmon")
plt.tick_params(labelbottom=False, labelleft=False)
plt.show
```

Paso 3: Definición de la arquitectura de la red neuronal.

Keras

```python
import tensorflow as tf
from tensorflow import keras as kr

from IPython.core.display import display, HTML
```

```python
lr = 0.01            #Learning rate
nn = [2, 16, 8, 1]   #número de neuronas por capa

# Creamos el objeto que contendrá a la red neuronal
# Secuencia de capas
model = kr.models.Sequential()

# Añadimos la capa 1:
l1 = model.add(kr.layers.Dense(nn[1], activation='relu'))

# Añadimos la capa 2:
l2 = model.add(kr.layers.Dense(nn[2], activation='relu'))

# Añaimos la capa 3:
l3 = model.add(kr.layers.Dense(nn[3], activation='sigmoid'))
```

```python
# Compilamos el modelo

model.compile(loss='mse', optimizer=kr.optimizers.SGD(learning_rate=0.05), metrics=['accuracy'])
```

```python
# Entrenamos el modelo
model.fit(X, Y, epochs=100)
```

Scikit Learn

```python
import sklearn as sk
import sklearn.neural_network
```

```python
lr = 0.01            #Learning rate
nn = [2, 16, 8, 1]   #número de neuronas por capa

# Creamos el objeto que contendrá a la red neuronal
clf = sk.neural_network.MLPRegressor(solver='sgd', learning_rate_init=lr, hidden_layer_sizes=tuple(nn[1:]), verbose=True, n_iter_no_change=1000, batch_size=100)
```

```python
# Entrenamos con nuestros datos
clf.fit(X, Y)
```

```python
# Obtener las predicciones:

Y_pred = clf.predict(X)
```

```python
# Calcular las métricas:
# Error Cuadrático Medio (MSE)
from sklearn.metrics import mean_squared_error as mse, r2_score

print(mse(Y, Y_pred))
```

```python
#Raíz del Error Cuadrático Medio (RMSE):

rmse = np.sqrt(mse(Y, Y_pred))
print(f"RMSE: {rmse}")
```

```python
#R-cuadrado
r2 = r2_score(Y, Y_pred)
print(f"R2: {r2}")
```
