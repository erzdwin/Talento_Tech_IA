## With Graphic

```python
import random as rd

nombres_peliculas = ["Spiderman", "Venom", "Batman", "Deadpool", "Avengers", "Avatar", "Titanic", "Gladiador", "Thor", "Joker", "Interstellar", "Transformers", "Matrix", "Ratatouille", "Alien", "Parasite", "Dune", "Jumanji", "Godzilla", "Scream"]

indice = rd.randint(0, len(nombres_peliculas) - 1)
movie = nombres_peliculas[indice]

Letra1 = movie[0]
Letrafn = movie[-1]

guion = " _ " * (len(movie) - 2)
Pista = [Letra1] + ['_'] * (len(movie) - 2) + [Letrafn]

print('Adivine la película: ', Pista)
Pista = list(Pista)

# Gráfico del ahorcado
ahorcado = [
    '''

     +---+
     |   |
         |
         |
         |
         |
    =========
    ''', '''
     +---+
     |   |
     O   |
         |
         |
         |
    =========
    ''', '''
     +---+
     |   |
     O   |
     |   |
         |
         |
    =========
    ''', '''
     +---+
     |   |
     O   |
    /|   |
         |
         |
    =========
    ''', '''
     +---+
     |   |
     O   |
    /|\  |
         |
         |
    =========
    ''', '''
     +---+
     |   |
     O   |
    /|\  |
    /    |
         |
    =========
    ''', '''
     +---+
     |   |
     O   |
    /|\  |
    / \  |
         |
    =========
    ''']

# Número de Intentos
intentos = 0
gano = False

while intentos < 7:

    entrada = input('Ingrese una letra o el nombre completo de la película: ')

    if len(entrada) == 1:
        if entrada in movie:
            print('La letra está en la película.')
            for i in range(len(movie)):
                if movie[i] == entrada:
                    Pista[i] = entrada
            print(''.join(Pista))
            if "_" not in Pista:
                gano = True
                break
        else:
            print('Incorrecto. La letra no está en la película.')
            print(ahorcado[intentos])
            intentos += 1
    else:
        if entrada == movie:
            gano = True
            break
        else:
            print('Incorrecto.')
            print(ahorcado[intentos])
            intentos += 1

if gano:
    print(" ╔═══════════╗")
    print(" ║  Ganaste!  ║")
    print(" ╚═══════════╝")
    print("   o ")
    print("  /|¡ ")
    print("  / \ ")
    print(" ¡Has ganado!")
else:
    print('Alcanzaste el número máximo de intentos.')
    print(" ╔═════════════════╗")
    print(" ║  GAME OVER :(  ║")
    print(" ╚═════════════════╝")
```
