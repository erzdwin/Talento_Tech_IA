```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

value_estudiante_list = ["Andres", "Maria", "Nicolas", "Alejandro", "Alex", "Ramiro", "Camila", "Erika", "Esteban", "Andrea"]

notas_estudiantes = {
    "Andres": [4.2, 3.9, 4.1],
    "Maria": [3.7, 3.5, 3.8],
    "Nicolas": [2.9, 3.5, 4.0],
    "Alejandro": [4.5, 4.6, 5.0],
    "Alex": [3.1, 4.0, 3.8],
    "Ramiro": [4.0, 3.5, 2.5],
    "Camila": [5.0, 3.0, 3.5],
    "Erika": [2.8, 1.0, 5.0],
    "Esteban": [3.5, 4.8, 3.0],
    "Andrea": [5.0, 4.5, 4.0]
}

value_promedio_list = []
for estudiante in value_estudiante_list:
    notas= notas_estudiantes[estudiante]
    if notas:
        promedio = round(sum(notas) / len(notas), 1)
        value_promedio_list.append(promedio)
    else:
        promedio = 0
        value_promedio_list.append(promedio)

value_asistencia_list = [100, 60, 50, 80, 95, 83, 100, 100, 50, 70] #en %

data_dict = {"Estudiante":value_estudiante_list,
             "Promedio de calificaciones":value_promedio_list,
             "Asistencia":value_asistencia_list}

df = pd.DataFrame(data_dict)

df

#Sistema de ingreso de notas

nombre_estudiante = input("Ingrese el nombre del estudiante: ")
nueva_nota = float(input("Ingrese la nueva nota: "))
nueva_asistencia = int(input("Ingrese el nuevo porcentaje de asistencia: "))

#Buscar al estudiante

indice_estudiante = df[df['Estudiante'] == nombre_estudiante].index

if indice_estudiante.empty:
    print("Estudiante no encontrado.")
else:
    indice_estudiante = indice_estudiante[0]

# Obtener las notas actuales del estudiante
notas_actuales = notas_estudiantes[nombre_estudiante]
# Agregar la nueva nota
notas_actuales.append(nueva_nota)
# Calcular el nuevo promedio
nuevo_promedio = round(sum(notas_actuales) / len(notas_actuales), 1)
# Actualizar el DataFrame
df.at[indice_estudiante, 'Promedio de calificaciones'] = nuevo_promedio
df.at[indice_estudiante, 'Asistencia'] = nueva_asistencia

df









```
